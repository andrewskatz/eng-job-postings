##### Drop duplicates

```python
def read_and_drop(file_name):
  raw_df = pd.read_csv(file_name)
  print(f"started with {raw_df.shape[0]} rows for {file_name}.")

  mod_df = raw_df.drop(columns='Unnamed: 0')
  mod_df = mod_df.drop_duplicates()

  print(f"ended with {mod_df.shape[0]} rows for {file_name}.")

  return mod_df
```
